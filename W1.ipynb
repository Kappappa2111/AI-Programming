#### Bài 1

```python
def is_prime(N):
    if N <= 1:
        return False
    for i in range(2, int(N**0.5) + 1):
        if N % i == 0:
            return False
    return True

number = int(input())
if is_prime(number):
    print(f"{number} là số nguyên tố")
else:
    print(f"{number} không là số nguyên tố")
```

#### Bài 2

```python
def is_prime(N):
    if N <= 1:
        return False
    for i in range(2, int(N**0.5) + 1):
        if N % i == 0:
            return False
    return True

def get_primes_in_range(A, B):
    primes = []
    for N in range(A, B + 1):
        if is_prime(N):
            primes.append(str(N))
    return primes

A = int(input("Nhập giá trị của A: "))
B = int(input("Nhập giá trị của B: "))

result = get_primes_in_range(A, B)
print(result)
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return abs(a * b) // gcd(a, b)

# Nhập giá trị của A và B từ người dùng
A = int(input("A = "))
B = int(input("B = "))

# Tính và in ra ước số chung lớn nhất (USCLN)
uscln = gcd(A, B)
print(f"Ước số chung lớn nhất của {A} và {B} là: {uscln}")

# Tính và in ra bội số chung nhỏ nhất (BSCNN)
bscnn = lcm(A, B)
print(f"Bội số chung nhỏ nhất của {A} và {B} là: {bscnn}")
```

```python
#5
def find_longest_words(word_squence):
  max_length = max(len(word) for word in word_squence)
  longest_words = [word for word in word_squence if len(word) == max_length]
  return longest_words

word_squence = input().split()
result = find_longest_words(word_squence)

if result:
  max_length = len(result[0])  # Lấy độ dài của từ dài nhất từ danh sách kết quả
  print(f"Từ dài nhất trong dãy là các từ có độ dài {max_length}:")
  for word in result:
    print(word)
```
